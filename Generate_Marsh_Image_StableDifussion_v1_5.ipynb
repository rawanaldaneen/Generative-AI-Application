```python
import torch
from diffusers import StableDiffusionPipeline
from PIL import Image

# 1. Load the "Standard" Model (Stable Diffusion v1.5)
# We use v1.5 because it is the standard benchmark for research papers.
model_id = "runwayml/stable-diffusion-v1-5"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe = pipe.to("cuda") # Use the GPU

# 2. Define your "Test Prompts" (These match your dataset)
prompts = [
    "A wooden Mashoof boat in the marshes",
    "A woman baking bread in a tradition clay Tanoor oven",
    "Water buffalo grazing in the marshlands",
    "A traditional reed Sarifa house",
    "Tradition Arab Marsh women wearing Dishdasha and Hijab"
]

# 3. Generate and Display the Images
print("Generating Baseline Images... (This proves the bias)")
for i, prompt in enumerate(prompts):
    image = pipe(prompt).images[0]

    # Save for your paper later
    filename = f"baseline_{i}.png"
    image.save(filename)

    print(f"\nPrompt: {prompt}")
    display(image)
    print("-" * 50)
```
